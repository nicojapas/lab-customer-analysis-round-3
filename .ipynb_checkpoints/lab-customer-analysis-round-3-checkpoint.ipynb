3. Show a plot of the total number of responses.

```python
def df_countplot(df, x, hue, y_label):
    sns.countplot(x=x, data=df, hue=hue)
    plt.ylabel(y_label)
    plt.show()
```

```python
df_countplot(data, 'response', 'sales_channel', 'Total number of responses')
```

```python
# Below a different way
```

```python
plot_y = [data['response'].value_counts().No, data['response'].value_counts().Yes]
plt.bar(data['response'].unique(), plot_y, color=['green', 'red'])
plt.show()
```

4. Show a plot of the response rate by the sales channel.

```python
# One way showing percentages
```

```python
response_no = []
response_yes = []
labels = []

def response_by_channel(x,y):
    for i in y.unique().tolist():
        response_total = (x[(y==i)].count())
        response_no.append(x[(y==i)].value_counts().No/response_total)
        response_yes.append(x[(y==i)].value_counts().Yes/response_total)
        labels.append(i)
        
response_by_channel(data['response'],data['sales_channel'])
```

```python
x_pos = np.arange(len(labels))
WIDTH = 0.4

plt.bar(x_pos, response_no, width=WIDTH)
plt.bar(x_pos+WIDTH, response_yes, width=WIDTH)
plt.xticks(x_pos+WIDTH/2, labels)

plt.show()
```

5. Show a plot of the response rate by the total claim amount.
